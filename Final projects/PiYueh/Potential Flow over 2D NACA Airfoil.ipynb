# Potential Flow Over A 2D NACA Airfoil Using Finite Difference Method

## Introduction

The potential flow theory has been used to study the flow over 2D airfoils for a long time. By complex analysis, we can obtain analytical solutions for many potential flows, especiall the potential flows over several kinds of 2D airfoils. After computers showed up in the world, people can also solve potential flows numerically using many numerical methods, like panel method, finite difference method, boundary element method etc.

Though there are many unrealistic assumption in potential flows, the potential flow theory provides a good beginning of further analyses. For example, many CFD software can solve potential flow first before they solve viscous flow in order to provide a good initial guess of viscous flow solver. 

In 2D potential flow, we can solve either potential function or stream function through their governing equtaions -- the Laplace equation. The governing equation of stream function is
$$
\nabla^2\Psi=\frac{\partial^2 \Psi}{\partial x^2}+\frac{\partial^2 \Psi}{\partial y^2}=0
$$
where $\Psi$ is stream function. If the velocity vector $\vec{V}=u\vec{i}+v\vec{j}$ represents the flow velocity, the relationship between flow velocity and the stream function is:
$$
u=\frac{\partial \Psi}{\partial y}\\
v=-\frac{\partial \Psi}{\partial x}
$$

In this notebook, we will show you how to solve potential flow over 2D airfoils using finite difference method.

## NACA 4-digit Airfoil Series

We use NACA 4-digit airfoils as our example in this notebook. Only the implementations are discussed here. For the introduction of NACA 4-digit airfoils, please refer to [reference [1]](http://en.wikipedia.org/wiki/NACA_airfoil).

Let's first define a Python function that can be used to generate the thickness profile of a NACA airfoil：

$$
y_t = 5t\left[ 0.2969\sqrt{\frac{x}{c}} -0.126\left(\frac{x}{c}\right) -0.3516\left(\frac{x}{c}\right)^2 +0.2843\left(\frac{x}{c}\right)^3 -0.1036\left(\frac{x}{c}\right)^4      \right]
$$

where $y_t$, $t$, $x$, and $c$ represent the thickness, maximum thickness, x-coordinate and chord length respectively. （Note: the last coefficient is modified from $-0.1015$ to $-0.1036$ in order to "close" the trailing edge. If we use $-0.1015$, there will be a thickness at trailing edge, just like airfoils in the real world. But in computational world, we prefer to use zero-thickness trailing edge.)

```python
def AFyt(x, t, c):
    '''
    input: 
        x: x coordinate of center line, float
        t: maximum thickness, in fraction of chord length, float
        c: chord lrngth, float
    output:
        half thickness of airfoil at corresponding x coordinate
    '''
    return 5. * t * (0.2969 * ((x/c)**0.5) - 
                     0.126 * (x/c) - 
                     0.3516 * ((x/c)**2) + 
                     0.2843 * ((x/c)**3) - 
                     0.1036 * ((x/c)**4))
```

$y_t$ is in fact just ***half thickness*** of the airfoil, so upper profile and lower profile should be generated separately. The lower part is negetive values of corresponding upper part. Using this function, we can generate the profiles for airfoils with the first and the second digits being zero (i.e. symmetrical airfoils). Here we show the profile of NACA 0014 for an example:

```python
x = np.linspace(0., 1., 101)
plt.plot(x, AFyt(x, 0.14, 1.0), 'k-', lw=2)    # upper surface
plt.plot(x, - AFyt(x, 0.14, 1.0), 'k-', lw=2)  # lower surface
plt.axis('equal'); plt.ylim((-0.5, 0.5))
```

For cambered 4-digit arifoils (i.e. the first and the second digit are not zero), we need to define Python functions for center lines (mean camber lines) and the angles, $\theta$, between center lines and the horizontal line.
$$
y_c = \left\{\begin{array}{lr}   
m\frac{x}{p^2}\left(2p-\frac{x}{c}\right)\text{,} && 0\le x\le p\times c\\
m\frac{c-x}{(1-p)^2}\left(1+\frac{x}{c}-2p\right)\text{,} && p\times c \le x \le c  
\end{array}\right.
$$

and

$$
\theta=\tan^{-1}{\frac{dy_c}{dx}}
$$

where

$$
\frac{dy_c}{dx}=\left\{\begin{array}{lr}
\frac{2m}{p^2}\left(p-\frac{x}{c}\right)\text{,} && 0\le x\le p\times c\\
\frac{2m}{(1-p)^2}\left(p-\frac{x}{c}\right)\text{,} && p\times c \le x \le c
\end{array}\right.
$$

```python
def AFyc(x, m, p, c):
    '''
    input:
        x: x coordinate of center line, float
        m: the maximum camber (100 m is the first of the four digits), float
        p: location of maximum camber (10 p is the second digit), float
        c: chord lrngth, float
    output:
        y coordinate of center line at corresponding x coordinate
    '''
    if (x >= 0.0) and (x <= p*c):
        return m * x * (2. * p - (x/c)) / (p**2.)
    elif (x > p*c) and (x <= c):
        return m * (c - x) * (1. + (x/c) - 2. * p) / ((1. - p)**2)
    else:
        raise ValueError
```

```python
def AFth(x, m, p, c):
    '''
    input:
        x: x coordinate of center line, float
        m: the maximum camber (100 m is the first of the four digits), float
        p: location of maximum camber (10 p is the second digit), float
        c: chord lrngth, float
    output:
        angle between center and horizontal line at corresponding x coordinate
    '''
    if (x >= 0.0) and (x <= p*c):
        return np.arctan(2.0 * m * (p - (x/c)) / (p**2))
    elif (x > p*c) and (x <= c):
        return np.arctan(2.0 * m * (p - (x/c)) / ((1. - p)**2))
    else:
        raise ValueError
```

Finally, the relationship between the coordinates on the upper and lower surface of cambered airfoils ($x_U$, $y_U$, $x_L$, and $y_L$), the center line coordinate ($x$ and $y_c$) and the thickness ($y_t$) are
$$
\begin{array}{ll}
x_U=x-y_t\sin{\theta}\text{,} && y_U=y_c+y_t\cos{\theta}\\
x_L=x+y_t\sin{\theta}\text{,} && y_L=y_c-y_t\cos{\theta}
\end{array}
$$

We wrote a function that can generate both symmetrical and cambered NACA 4-digit airfoils and use a sign to represent upper or lower profile:

```python
def AF(x, t, sign, m, p, c):
    '''
    input:
        x: x coordinate of center line, float
        t: maximum thickness, in fraction of chord length, float
        sign: indicate upper (1) or lower (-1) surface of airfoil
        m: the maximum camber (100 m is the first of the four digits), float
        p: location of maximum camber (10 p is the second digit), float
        c: chord lrngth, float
    output:
        x, y coordinates on airfoil surface at corresponding 
        center line x coordinate
    '''
    if (m == 0.) or (p == 0):
        return x, sign * AFyt(x, t, c)
    else:
        return np.array([x[i] - 
                         sign * AFyt(x[i], t, c) * np.sin(AFth(x[i], m, p, c)) 
                         for i in range(np.size(x))]), \
               np.array([AFyc(x[i], m, p, c) + 
                         sign * AFyt(x[i], t, c) * np.cos(AFth(x[i], m, p, c))
                         for i in range(np.size(x))])
        
```

The following shows the profile of a NACA 2412 airfoil for an example.

```python
x = np.linspace(0., 1., 101)
xU, yU = AF(x, 0.12, 1, 0.02, 0.4, 1.0)
xL, yL = AF(x, 0.12, -1, 0.02, 0.4, 1.0)
plt.plot(xU, yU, 'k-', lw=2)
plt.plot(xL, yL, 'k-', lw=2)
plt.axis('equal'); plt.ylim((-0.5, 0.5))
```

## Discretization in Space of NACA 2412 Airfoil

### Physical Domain and Computational Doamin

Here we introduce the concept of physical and computational domains follow the chapter 9 in reference [2]. 

Apparently, when we are using finite difference method, ractangular grids are best. However, if there are objects inside our domain, for example, airfoils, it is difficult to discretize the domain into rectangular grids. 

Such domain can be discretized into structured mesh. Such mesh, however, may contain arbitrary quadrilateral grids, instead of rectangular grids. So the idea is to map our real domain to another space that can be discretized into rectangular grids and solve our problems in this space. The real domain is called physical domain, and the space in which we solve our problems is called computational domain. The following figure shows the concept of physical and computational domain.
![Fig. 1](./DomainIllust.jpg)

We define the coordinate in computational domain to be $\xi$ and $\eta$ (called computational coordinate) and that in physical doamin to be $x$ and $y$ (physical coordinate). Physical coordinates are functions of computational coordinates, and vice versa. This means that all calculations performed in physical domains can also be performed in computational domain through the relationship between $x$, $y$ and $\xi$, $\eta$ after their relationship has been determined. 

If we fix the the computational coordinates of grid points in a way that all grids in computational domain are rectangular grids with $\Delta\xi=\Delta\eta=1$, what we need to do next is to determine the physical coordinates of these grid points in the computational domain. Let $i$ and $j$ be the index of grid points in computational domain. Then, $x_{i,j}$/$y_{i,j}$ represents the x/y-component of physical coordinate of the grid points $(i,j)$ in the computational domain. Then $x_{i,j}$/$y_{i,j}$ is what we want.

In this notebook, we use O-type mesh on physical domain, that is, the outer boundary of physical domain is a circle, shown as below. 
![Fig. 2](./PDomain.jpg)

We devide the annular area between the arifoil and outer boundary by a line $\overline{AC}$:
![Fig. 3](./Devide.jpg)

and then "expand" the devided annular area to a rectangular domain, which will be our computational domain:
![Fig .4](./CDomain.jpg)

From the name of boundaries ($B_1$~$B_4$), we can understand the relationship between our physical domain and the computational domain.

### Grid Generation

The grids on computational domain are simply rectangular grids. The lengths of boundaries and the size of the grids in the computational domain are not important, because they only make difference on the transformation relationships between two domains. So let's fix the size of grids in our computational domain as 1, i.e. $\Delta\xi=\Delta\eta=1$. Also, the numbers of nodes on $\xi$ and $\eta$ directions are $N_{\xi}=51$ and $N_{\eta}=21$. The grids on computational domain are:

```python
Nxi = 51
Neta = 21
eta, xi = np.meshgrid(np.linspace(0, Neta-1, Neta), np.linspace(0, Nxi-1, Nxi))
```

```python
def plotMesh(x, y):
    for i in range(Nxi):
        plt.plot(x[i, :], y[i, :], 'k.-', lw=2)
    for i in range(Neta):
        plt.plot(x[:, i], y[:, i], 'k.-', lw=2)
```

```python
plotMesh(xi, eta)
```

The next step is to determine the physical coordinates of the grid points, $x_{i,j}$ and $y_{i,j}$, in the computational domain. In the following two subsections, we introduce two grid generation methods: the algebraic method and elliptic method.

#### Algebraic Grid Generation

The concept of algebraic grid generation is to interpolate the physical coordinates of interior grid points between two known boundaries, i.e. the airfoil and the outer boundary in our case.

In our case, from the illustration in the previous content, we know the numbers of grids on the airfoil and outer boundary should be the same with $N_{\xi}$, while that on the dividing line, $\overline{AC}$, should equal $N_{\eta}$. We also know the physical coordinates of grids on $\eta=0$ and $\eta=N_{\eta}$, because they are located on the airfoil surface and the outer boundary and can be determined by us, that is 

$$
\begin{array}{ll}
x_{i, j=0} = x_{airfoil}\text{,} && y_{i, j=0} = y_{airfoil}\\
x_{i, j=-1} = x_{outer\ boundary}\text{,} && y_{i, j=-1} = y_{outer\ boundary}
\end{array}
$$

For convenience, we here use Python index, where j=-1 represents the last grid points in $\eta$ direction.

The x and y coordinates for interior points (i.e. $0\le i\le -1$ and $1\le j \le -2$, using Python index) can be obtained through interpolation. This is called algrbraic grid generation method.

We now generate algebraic grids for a NACA 2412 airfoil. The order of grid points on airfoil is clockwise and starts from the trailing edge. The center of circular outer boundary is located at $0.5c$.

```python
rBC = 5.0    # the radius of outer boundary
m, p, t, c = 0.02, 0.4, 0.12, 1.0    # parameters of NACA 2412 airfoil

# Initialize x[i, j] and y[i, j]
x = np.empty((Nxi, Neta))
y = np.empty((Nxi, Neta))

# Generate grid points on airfoil surface
Nxc = (Nxi-1)/2 + 1
xc = np.linspace(0., 1., Nxc)
xU, yU = AF(xc, 0.12, 1, 0.02, 0.4, 1.0)
xL, yL = AF(xc, 0.12, -1, 0.02, 0.4, 1.0)

# Set x_{i, j=0} and y_{i, j=0}
x[:Nxc, 0] = xL[-1::-1].copy()
x[Nxc:, 0] = xU[1:].copy()

y[:Nxc, 0] = yL[-1::-1].copy()
y[Nxc:, 0] = yU[1:].copy()

# Generate grid points on circular outer boundary
# and set x_{i, j=-1}, y_{i, j=-1}
dr = 2. * np.pi / (Nxi -1)
th = - np.array([i * dr for i in range(Nxi)])
x[:, -1] = rBC * np.cos(th) + 0.5 * c
y[:, -1] = rBC * np.sin(th)
```

Interpolate the coordinates of interior grid points.

```python
for i in range(Nxi):
    x[i, 1:-1] = np.linspace(x[i, 0], x[i, -1], Neta)[1:-1]
    y[i, 1:-1] = np.linspace(y[i, 0], y[i, -1], Neta)[1:-1]
```

Let's see how the mesh looks like.

```python
plt.figure(figsize=(8, 8), dpi=100)
plotMesh(x, y)
plt.axis('equal')
plt.xlim((-4.5, 5.5)); plt.ylim((-5, 5))
```

Look closer to the vicinity of the airfoil.

```python
plt.figure(figsize=(8, 8), dpi=100)
plotMesh(x, y)
plt.axis('equal')
plt.xlim((-0.5, 1.5)); plt.ylim((-1, 1))
```

Though the mesh quality wasn't actually examined here, it looks not good near the airfoil. That's why we need other grid generations method.

#### Elliptic Grid Generation

In elliptic grid generation method, we use elliptic PDEs to constrain the relationship between physical and computational coordinates. The elliptic PDEs used in this grid generation method are:

$$
\frac{\partial^2 \xi}{\partial x^2}+\frac{\partial^2 \xi}{\partial y^2}=0\\
$$
$$
\frac{\partial^2 \eta}{\partial x^2}+\frac{\partial^2 \eta}{\partial y^2}=0\\
$$

However, if we solve these PDEs (if we could), what we will obtain are the computational coordinates, i.e. $\xi=\xi(x, y)$ and $\eta=\eta(x, y)$, which are not what we want, because we already fixed the computational coordinates (they're simply rectangular grids with $\Delta\xi=\Delta\eta=1$).

Therefore, the actual PDEs we are going to solve, which are derived from the above PDEs (see reference [2]), are

$$
a\frac{\partial^2 x}{\partial \xi^2}-2b\frac{\partial^2 x}{\partial \xi \partial \eta}+c\frac{\partial^2 x}{\partial \eta^2}=0
$$

$$
a\frac{\partial^2 y}{\partial \xi^2}-2b\frac{\partial^2 y}{\partial \xi \partial \eta}+c\frac{\partial^2 y}{\partial \eta^2}=0
$$

where

$$
a=\left(\frac{\partial x}{\partial \eta}\right)^2+\left(\frac{\partial y}{\partial \eta}\right)^2\\
b=\left(\frac{\partial x}{\partial \xi}\right)\left(\frac{\partial x}{\partial \eta}\right)+\left(\frac{\partial y}{\partial \xi}\right)\left(\frac{\partial y}{\partial \eta}\right)\\
c=\left(\frac{\partial x}{\partial \xi}\right)^2+\left(\frac{\partial y}{\partial \eta}\right)^2
$$

We can see that, in order to solve the PDEs, we need two boundary conditions on each direction, This means we need $x_{i,j=0}$, $x_{i,j=-1}$, $x_{i=0,j}$, $x_{i=-1,j}$, $y_{i,j=0}$, $y_{i,j=-1}$, $y_{i=0,j}$, and $y_{i=-1,j}$.

Luckly, $x_{i,j=0}$, $x_{i,j=-1}$, $y_{i,j=0}$, and $y_{i,j=-1}$ are known and are physical coordinates of nodes on the airfoils and outer boundary, as described in the previous subsection.

From the illustration in the previous content, we also know that $x_{i=0,j}$ and $x_{i=-1,j}$ are actually the same, because they both represent the physical coordinates of nodes on dividing line $\overline{AC}$. $y_{i=0,j}$ and $y_{i=-1,j}$ are the same, too. This implies ***periodic boundary conditions*** can be applied on $x_{i=0,j}$, $x_{i=-1,j}$, $y_{i=0,j}$, and $y_{i=-1,j}$.

These PDEs are non-linear. Iterative methods can be used to solve them. An initial guess is necessary for iterative methods. The results obtained from algebraic grid generation method can be a good initial guess. The basic Jacobi iterative method is used here.

We can use central difference on these PDEs. And because $\Delta\xi=\Delta\eta=1$, the discretized equations are very simple. Since the forms of these two PDEs are the same, their discretized PDEs are also the same. The discretized equation of $x$ are

$$
a(x_{i+1, j}-2x_{i,j}+x_{i-1, j}) - \frac{1}{2}b(x_{i+1, j+1}-x_{i+1, j-1}+x_{i-1,j-1}-x_{i-1,j+1})+c(x_{i, j+1}-2x_{i,j}+x_{i, j-1})=0
$$

or

$$
x_{i,j}=\frac{1}{2}\left\{
a(x_{i+1, j}+x_{i-1, j}) - \frac{1}{2}b(x_{i+1, j+1}-x_{i+1, j-1}+x_{i-1,j-1}-x_{i-1,j+1})+c(x_{i, j+1}+x_{i, j-1})
\right\} / (a+c)
$$

where

$$
a=(\frac{x_{i,j+1}-x_{i,j-1}}{2})^2+(\frac{y_{i,j+1}-y_{i,j-1}}{2})^2\\
b=(\frac{x_{i+1,j}-x_{i-1,j}}{2})(\frac{x_{i,j+1}-x_{i,j-1}}{2})+(\frac{y_{i+1,j}-y_{i-1,j}}{2})(\frac{y_{i,j+1}-y_{i,j-1}}{2})\\
c=(\frac{x_{i+1,j}-x_{i-1,j}}{2})^2+(\frac{y_{i+1,j}-y_{i-1,j}}{2})^2
$$

Replace $x$ with $y$ when we want to solve $y$.

We can solve these PDEs now. The criterion for the iteration to stop is that the maximum difference between current result and the result of the last iteration is less than $10^{-6}$.

```python
def Solve_a_b_c(x, y):
    '''
    input:
        x: the x coordinate of x_{i=i-1~i+1, j=j-1~j+1}, at least 3x3 array
        y: the y coordinate of y_{i=i-1~i+1, j=j-1~j+1}, at least 3x3 array
    output:
        a, b, c: at least 1x1 float
    '''
    a = 0.25 * (((x[1:-1, 2:] - x[1:-1, :-2])**2) + 
                ((y[1:-1, 2:] - y[1:-1, :-2])**2))
    b = 0.25 * ((x[2:, 1:-1] - x[:-2, 1:-1]) * 
                (x[1:-1, 2:] - x[1:-1, :-2]) + 
                (y[2:, 1:-1] - y[:-2, 1:-1]) * 
                (y[1:-1, 2:] - y[1:-1, :-2]))
    c = 0.25 * (((x[2:, 1:-1] - x[:-2, 1:-1])**2) + 
                ((y[2:, 1:-1] - y[:-2, 1:-1])**2))
    return a, b, c
```

```python
def SolveEq(a, b, c, U):
    '''
    input:
        a, b, c: as described in the content
        U: the result of the last iteration
    output:
        return the result of current iteration
    '''
    return 0.5 * (
                  a * (U[2:, 1:-1] + U[:-2, 1:-1]) + 
                  c * (U[1:-1, 2:] + U[1:-1, :-2]) -
                  b * 0.5 * (U[2:, 2:] - U[2:, :-2] + U[:-2, :-2] - U[:-2, 2:])
                 ) / (a + c)
```

```python
iters=0
while True:
    
    # count the number of iterations
    iters += 1
    
    # backup the last result
    xn = x.copy()
    yn = y.copy()
    
    # solve periodic BC first
    tempx = np.append([x[-2, :].copy()], x[0:2, :].copy(), 0) 
    tempy = np.append([y[-2, :].copy()], y[0:2, :].copy(), 0)
    a, b, c = Solve_a_b_c(tempx, tempy)
    x[0, 1:-1] = SolveEq(a, b, c, tempx)
    y[0, 1:-1] = SolveEq(a, b, c, tempy)

    x[-1, 1:-1] = x[0, 1:-1].copy()
    y[-1, 1:-1] = y[0, 1:-1].copy()
    
    # solve interior
    a, b, c = Solve_a_b_c(x, y)
    x[1:-1, 1:-1] = SolveEq(a, b, c, x)
    y[1:-1, 1:-1] = SolveEq(a, b, c, y)
    
    # calculate difference between current and the last result
    errx = np.abs(x - xn)
    erry = np.abs(y - yn)
    
    # adjudge whether the iteration should stop
    if (errx.max() <= 1e-6) and (erry.max() <= 1e-6):
        break
```

Let's see how the mesh looks like.

```python
plt.figure(figsize=(8, 8), dpi=100)
plotMesh(x, y)
plt.axis('equal')
plt.xlim((-4.5, 5.5)); plt.ylim((-5, 5))
```

The grids near the airfoil look more clustered. This is good. And how about the grids in the vicinity of the airfoil?

```python
plt.figure(figsize=(8, 8), dpi=100)
plotMesh(x, y)
plt.axis('equal')
plt.xlim((-0.5, 1.5)); plt.ylim((-1, 1))
```

## Potential Flow over NACA 2412

### Governing Equation in the Computational Domain

We know the governing equation of the stream function, $\Psi$, in the physical domain is the Laplace equation, as described in the introduction section. To solve the stream function in the computational domain, we need the governing equation for that. According to the derivation described in the reference [3], we can obtain the governing equation for stream function in the computational domain:

$$
a\frac{\partial^2 \Psi}{\partial \xi^2}-2b\frac{\partial^2 \Psi}{\partial \xi \partial \eta}+c\frac{\partial^2 \Psi}{\partial \eta^2}=0
$$

where

$$
a=\left(\frac{\partial x}{\partial \eta}\right)^2+\left(\frac{\partial y}{\partial \eta}\right)^2\\
b=\left(\frac{\partial x}{\partial \xi}\right)\left(\frac{\partial x}{\partial \eta}\right)+\left(\frac{\partial y}{\partial \xi}\right)\left(\frac{\partial y}{\partial \eta}\right)\\
c=\left(\frac{\partial x}{\partial \xi}\right)^2+\left(\frac{\partial y}{\partial \eta}\right)^2
$$

The form of the governing equation is the same with waht we solved in elliptic grid generation method! This means we can use the functions we previously wrote for elliptic grid generation method!

### Boundary Conditions in the Computational Domain

Though the form of the governing equation is the same with PDEs in elliptic grid generation method, the boundary conditions are not. We now examine the boundary conditions. 

For the outer boundary, we expect the flow velocity on this boundary is equal to free stream velocity, that is

$$
\left\{
\begin{array}{l}
u=\frac{\partial \Psi}{\partial y}=V_\infty\cos(AOA)\\
v=-\frac{\partial \Psi}{\partial x}=V_\infty\sin(AOA)
\end{array}\right.,\ on\ outer\ boundary
$$

where $AOA$ represents angle of attact, and $V_\infty$ is free stream velocity.

Since $V_\infty$ and $AOA$ are constant, we can simply integrate the above boundary conditions for the stream function on the outer boundary

$$
\Psi=-xV_\infty\sin(AOA)+yV_\infty\cos(AOA)+C_1,\ on\ outer\ boundary
$$

In potential flow, the values of stream function are not important. The important matter is the derivatives of the stream function. As the constant $C_1$ in the above equation will disappear in derivatives, we can simply make $C_1=0$ here. Now we can apply Dirichlet boundary condition on the outer boundary:

$$
\Psi_{i,j=-1}=-x_{i,j=-1}V_\infty\sin(AOA)+y_{i,j=-1}V_\infty\cos(AOA)
$$

Next, we examine the boundary conditions for nodes on the airfoil. For walls in potential flows, no-slip boundary condition is invalid due to the lack of viscosity. The boundary condition for walls in potential flows is that the flow can not penetrate the wall. This implies that the wall itself is a streamline.

$$
\Psi=C_2,\ on\ the\ airfoil
$$

or

$$
\Psi_{i,j=0}=C_2
$$

Since we have determined the values of stream function on outer boundary, we can not determine the value of stream function on the airfoil, $C_2$, now (because we don't know which streamline on the outer boundary corresponds to the streamline on the airfoil surface). Fortunately, we will reslove this problem later after we introduce the Kutta condition.

Finally, the boundary condition on dividing line $\overline{AC}$ is again periodic boundary condition.

$$
\Psi_{i=0,j}=\Psi_{i=-1,j}
$$

### The Kutta Condition

In potential flow, due to the lack of viscosity, the flow under airfoil will flow back to the upper surface after it passed by the trailing edge even in a low angle of attact and low free stream velocity. This is not physical. The Kutta condition can ensure that situation won't happen. The Kutta condition makes the flow smoothly passes by trailing edges. More details can be found in many textbook of inviscid flow or potential flow. Here we only introduce the implementation of the Kutta condition in our code.

According to the reference [3], for the airfoils with finite-angle trailing edges, like NACA 2412, the Kutta condition gives us that the value of stream function on the node next to trailing edge should be the same with that on the trailing edge, in order to provide a smooth streamline near the trailing edge. In other words,

$$
\Psi_{i=0, j=1}=\Psi_{i=0, j=0}
$$

Recall the boundary condition on the airfoil surface, $\Psi_{i,j=0}=C_2$. We can determine $C_2$ and $\Psi_{i,j=0}$ now. In each iteration when solving the governing equation, let

$$
\Psi_{i,j=0}=C_2=\Psi_{i=0,j=1}
$$

The problem of the boundary condition on the airfoil surface is then resolved. We can start to solve the stream functions! 

### Solve It Now!!!

Let the angle of attact be $15^\circ$ and free stream velocity be $70\ m/s$. Other parameters are the same with what we used in previous sections.

```python
# set angle of attact
AOA = 15. / 180. * np.pi

# set free stream velocity
Vinf = 70.

# initialize stram functions
stream = np.zeros((Nxi, Neta))

# set up the BCs on the outer boundary
stream[:, -1] = - x[:, -1] * Vinf * np.sin(AOA) + y[:, -1] * Vinf * np.cos(AOA)

# solve the PDE by iterative method
iters = 0
while True:
    
    # count the number of current interation
    iters += 1
    
    # backup the last result
    stream_n = stream.copy()
    
    # apply periodic BC on dividing line
    temp = np.append([stream[-2, :].copy()], stream[:2, :].copy(), 0)
    tempx = np.append([x[-2, :].copy()], x[:2, :].copy(), 0) 
    tempy = np.append([y[-2, :].copy()], y[:2, :].copy(), 0)
    a, b, c = Solve_a_b_c(tempx, tempy)
    stream[0, 1:-1] = SolveEq(a, b, c, temp)
    stream[-1, :] = stream[0, :].copy()
    
    # apply Kutta condition 
    # and set the value of stream function on the airfoil surface
    stream[:, 0] = stream[0, 1]

    # solve interior
    a, b, c = Solve_a_b_c(x, y)
    stream[1:-1, 1:-1] = SolveEq(a, b, c, stream)
    
    # calculate difference between current and the last result
    err = np.abs(stream - stream_n)
    
    # adjudge whether the iteration should stop
    if (err.max() <= 1e-6):
        break
```

Though the values of stream function are not important, conventionally, we make the stramline on the airfoil surface to be zero.

```python
stream = stream - stream[0, 0]
```

Finally, the streamlines are shown in the following figure.

```python
# set the contour lines with negative values to be solid lines
import matplotlib as mpl
mpl.rcParams['contour.negative_linestyle'] = 'solid'

# contour
plt.figure(figsize=(10, 8), dpi=100)
cs = plt.contour(x, y, stream, 100, colors='k')
plt.clabel(cs)
plt.plot(x[:, 0], y[:, 0], 'k-', lw=1) # plot the airfoil
plt.xlim((-2., 3.))
plt.ylim((-1.5, 1.5))
```
